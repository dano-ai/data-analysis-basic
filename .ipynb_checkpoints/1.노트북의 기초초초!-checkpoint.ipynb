# Cell이란 무엇인가?

## Cell 유형은 3가지다.
- Code
- Markdown
- Raw (몰라도 된다.)

### Code는 파이썬 코드를 입력하고 실행한다.
셀을 만들면 일단 Code 유형으로 만들어진다.
코드를 입력하고 Shift+Enter를 치면 코드가 실행된다.

```python
ls
```

ls는 현재 폴더에 있는 파일과 하위 폴더를 보여준다.

```python
코드 셀에 주석을 입력하려면 '#'을 입력하고 원하는 내용을 적으면 된다.
```

```python
ls # ls는 현재 폴더에 있는 파일과 하위 폴더를 보여준다.
```

### Markdown (마크다운)은 메모장이다. 
원하는 글 아무거나 적을 수 있다. 
code 유형인 셀을 Markdown으로 변경하고 글을 적어 보자. (상단 메뉴 이용해서)

마찬가지로 글 다 적고 나서 Shift+Enter를 치면 된다.

상단 메뉴를 이용해서 Cell을 추가할 수 있다. Insert 셀 위에/아래에

# 키보드 숏컷을 배우자
키보드 숏컷을 배우면 더 빠르게 작업할 수 있다.
Help > Keyboard shortcuts

## 셀 선택하기

셀에서 입력하다가 ESC 키를 누르면 해당 셀을 선택할 수 있다. 마우스로 셀 앞쪽을 클릭한 것과 동일한 상태가 된다.

이 상태에서 화살표 키를 누르면 위/아래 셀로 이동할 수 있다.

## 셀 추가 삭제
하나의 셀을 선택하고

a: Above. 셀 위에 추가
b: Below. 셀 아래에 추가    
dd: 셀 삭제. 실수로 잘못 지우면 Edit 메뉴에서 Undo Delete Cells로 복구.

## 셀 유형 변경하기
셀을 마우스로 선택하고

m: markdown으로 변경
y: code로 변경

markdown에서 code로, 다시 code에서 markdown으로 변경해 보자.

```python
ls
```

## 셀 Copy & Paste
셀을 마우스로 선택하고

c: 셀 전체가 복사된다.
    
v: 현재 커서가 선택한 셀 아래에 복사한 셀을 붙여넣기한다.

# Markdown 활용하기

## 제목 서식 만들기
markdown cell에서 '#'을 붙이면 헤더 (제목)을 만들 수 있다. '##', '###'도 가능하다.

**주의: '#' 치고 꼭 한칸 띄워야 한다.**

# 1
## 2
### 3
#### 4

## font 스타일 적용하기

'*'을 앞뒤로 입력하면 글자 스타일이 바뀐다. 공백 없이 붙여서 앞과 뒤에 똑같은 개수를 붙여라.

*기울임*

**볼드체로**

***볼드이면서 기울이기***
