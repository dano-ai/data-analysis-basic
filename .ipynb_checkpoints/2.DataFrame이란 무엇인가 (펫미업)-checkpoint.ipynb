## CSV 화일 불러오기
csv 화일을 읽어서 dataFrame에 저장합니다.

    pd.read_csv(화일위치)
    
header (컬럼이름)이 한줄 있는 경우 아래와 같이 사용합니다.

    pd.read_csv(화일위치, header=1)

```python
df = pd.read_csv('data/펫미업 - 6개월치 예약목록 - 20181029.csv', header=1)
```

## Group해서 보기

```python
df.columns
```

```python
df['고객등급']
```

```python
df.groupby('고객등급').count()['예약코드'].sort_values(ascending=False)
```

### 퍼센티지 구하기

```python
(df.groupby('고객등급').count()['예약코드'] / df.shape[0] * 100).sort_values(ascending=False)
```
