## CSV 화일 불러오기
csv 화일을 읽어서 dataFrame에 저장합니다.

    pd.read_csv(화일위치)
    
header (컬럼이름)이 한줄 있는 경우 아래와 같이 사용합니다.

    pd.read_csv(화일위치, header=1)
