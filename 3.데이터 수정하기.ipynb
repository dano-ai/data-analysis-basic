```python
df = pd.read_csv('data/펫미업 - 6개월치 예약목록 - 20181029.csv', header=1)
```

## 빈 값 채우기

NaN을 어떻게 할 것인가?

    dataframe.fillna(채우고자하는값)

```python
df['방문경로']
```

```python
df['방문경로'] = df['방문경로'].fillna("모름")
```

```python
df['방문경로']
```

```python
df.groupby(['방문경로']).count()['예약코드']
```

```python
df.groupby(['방문경로']).count()['예약코드'] / df.shape[0] * 100
```

## 일자와 시간 분리하기

### 시각이 문자열로 들어와 있는 경우

예. 접수일자
2018-03-27 19:35

```python
from datetime import datetime
```

```python
df['접수일자'][0:1]
```

```python
df = df.rename(columns={'접수일자': '접수TimeStamp'})
```

```python
df['접수TimeStamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%Y-%m-%d'))
```

```python
df['접수일자'] = df['접수TimeStamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%Y-%m-%d'))
```

```python
df['접수TimeStamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%H:%M'))
```

```python
df['접수시각'] = df['접수TimeStamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%H:%M'))
```

```python
df.columns
```

```python
df['출발지']
```

```python
df_depart_from_seoul = df[df['출발지'].apply(lambda x: x.startswith("서울"))]
```

```python
df_depart_from_seoul
```

```python
import re # 문자열에서 패턴을 찾을 때 사용한다.

# 참고 https://wikidocs.net/4308#findall
```

```python
df_depart_from_seoul.shape
```

```python
df_depart_from_seoul['출발지'].apply(lambda x: re.findall(' (.+?)구', x)[0]+"구")
```

```python
df_depart_from_seoul['출발지(구)'] = df_depart_from_seoul['출발지'].apply(lambda x: re.findall(' (.+?)구', x)[0]+"구")
```

```python
df_depart_from_seoul['도착지(광역)'] = df_depart_from_seoul['도착지'].apply(lambda x: re.findall('(.+?) ', x)[0])
```

```python
df_depart_from_seoul.groupby(['도착지(광역)']).count()['예약코드']
```

#### 퍼센티지 계산하기

```python
df_depart_from_seoul.groupby(['도착지(광역)']).count()['예약코드'].apply(lambda x: x / df_depart_from_seoul.shape[0] *100).sort_values(ascending=False)
```

```python
df_depart_from_seoul.columns
```

```python
df_depart_from_seoul.groupby(['출발지(구)']).count()['예약코드'].apply(lambda x: x / df_depart_from_seoul.shape[0] *100).sort_values(ascending=False)
```

## 도착지에 병원 이름 있는지 찾기

```python
df_depart_from_seoul['도착지'].apply(lambda x: False if not re.findall('병원', x) else True).sum()
```

```python
df_depart_from_seoul['병원여부'] = df_depart_from_seoul['도착지'].apply(lambda x: False if not re.findall('병원', x) else True)
```

```python
df_depart_from_seoul['병원여부'].sum() / df_depart_from_seoul.shape[0] * 100
```
